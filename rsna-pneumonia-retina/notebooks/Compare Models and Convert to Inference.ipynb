```python
%matplotlib inline 
import os
from glob import glob 
import pandas as pd
import matplotlib.pyplot as plt 

snapshots_path  = '../snapshots_fulltraining/'

## Listing all h5 files not converted as inference
models = [f for f in glob(f'{snapshots_path}*.h5') if 'infer' not in f]

## Listing the first result file in the directory 
results = [glob(f'{snapshots_path}*.csv')][0][0]

models, results
```

## Selecting the best 5 models given rsna_score

```python
df_results = pd.read_csv(results)
best_rsna_scores = df_results.sort_values('y_rsna_score', ascending=False).head(5)
best_rsna_scores_models = best_rsna_scores['Unnamed: 0'].values
best_rsna_scores
```

## Selecting the 5 best Loss models 

```python
best_loss = df_results.sort_values('loss', ascending=True).head(5)
best_loss_models = best_loss['Unnamed: 0'].values
best_loss
```

```python
best_models_intersection = [m for m in best_rsna_scores_models if m in best_loss_models]
best_models_intersection
```

## Selecting the best models by classification loss

```python
best_classification_loss = df_results.sort_values('classification_loss', ascending=True).head(5)
best_classification_loss_models = best_loss['Unnamed: 0'].values
best_classification_loss
```

```python
best_models_intersection = [m for m in best_models_intersection if m in best_classification_loss_models]
best_models_intersection
```
