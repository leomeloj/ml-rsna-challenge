```python
import visualize_prediction as V
import eval_model as E
import pandas as pd
from torchvision import transforms, utils

#suppress pytorch warnings about source code changes
import warnings
warnings.filterwarnings('ignore')
```

```python
PATH_TO_IMAGES = "../../test/data/"
PATH_TO_MODEL = "results/checkpoint"
```

```python
dataloader,model= V.load_rsna_test_data(PATH_TO_IMAGES,PATH_TO_MODEL)
print(model)
print(len(dataloader))
```

```python
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

data_transform = transforms.Compose([
    transforms.Scale(224),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])

model.cuda()
    
pred = E.predict_test(data_transform, model, PATH_TO_IMAGES)
```

```python
print(pred)
```

```python
ths_prob = 0.5817979 #0.51566422 #0.55670476
df_positive = pred[pred['prob_Pneumonia'] > ths_prob]
df_to_predict = df_positive['patientId']
df_to_predict.to_csv('test_to_predict.csv')
```
